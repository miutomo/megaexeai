```python
import glob
#globは正規表現でファイル検索を可能にするモジュール、基本では入らないので別途インストールが必要
import os
import shutil
```

```python
path = os.getcwd()
print(path)
#glob.glob(path)
```

```python
flist = glob.glob('**/*.jpg')
print(type(flist),len(flist))
```

```python
#フォルダがなければ作成
src_dir = 'allphoto'
if os.path.exists(src_dir) != True:
    tmpath = os.path.join(path, src_dir)
    os.mkdir(tmpath)
else:
    pass
```

```python
#ファイルリスト分繰り返して移動（コピー）
for i, name in enumerate(flist):
    tmp_path = os.path.split(name)[1]
    dst_path = os.path.join(src_dir, tmp_path)
    
    shutil.copy2(name, dst_path)

#最後に処理完了を表示
print('finished!')
```

```python
#確認用
os.chdir(src_dir)
cnt = glob.glob('*.jpg')

for no, name in enumerate(cnt):
    print(no, name)
    
print(len(cnt))
```
